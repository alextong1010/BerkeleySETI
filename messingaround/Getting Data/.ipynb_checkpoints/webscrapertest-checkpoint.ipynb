```python
import requests
import numpy as np
from bs4 import BeautifulSoup
```

```python
URL = "http://tessebs.villanova.edu/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
#print(soup.prettify())
test = soup.find_all('a')
print(test)
```

```python
for t in test:
    text = t.text.strip()
    if text.isnumeric():
        print(text)
```

```python
x = np.empty((4584,2), int)
```

```python
row = 0
for t in test:
    text = t.text.strip()
    if text.isnumeric():
        x[row][0] = text
        row=row+1
```

```python
print(x)
```

```python
html = soup.find_all('td')
print(html)
```

```python
x = np.empty((4584,2), dtype='object')
index = 0
for t in html:
    text = t.text.strip()
    if (index%9==0 or index%9==1):
        x[int(index/9)][index%9]=text
    index=index+1


        
```
